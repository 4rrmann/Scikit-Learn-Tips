***Pruning of decision trees to avoid overfiting!***
- *uses cost-complexity pruning*
- *increase `ccp_alpha` to increase pruning (default values is 0)*

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

sdf = df[['age', 'pclass', 'fare', 'survived']]
sdf = sdf.dropna()

X = sdf.drop('survived', axis=1)
y = sdf['survived']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
#default tree has 491 nodes
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X,y).tree_.node_count
```

```python
cross_val_score(dt, X, y, cv=5, scoring='accuracy').mean()
```

```python
#pruned tree 11 nodes
dt = DecisionTreeClassifier(
    ccp_alpha=0.01, #inc-Alpha to inc-Pruning
    random_state=0
)

dt.fit(X, y).tree_.node_count
```

```python
#pruning improved the cross-validation accuracy
cross_val_score(dt, X,y, cv=5, scoring='accuracy').mean()
```
