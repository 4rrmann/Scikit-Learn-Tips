*need to get the faature names output by a ColumnTransformer?*

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df

X = df[['embarked', 'who', 'parch', 'fare']].dropna()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#one-hot encoded "embarked" & "who", and passthrough "parch" & "fare"
ct = make_column_transformer(
    (OneHotEncoder(), ['embarked', 'who']),
    remainder='passthrough'
)
```

```python
#CT outputs 7 columns:
ct.fit_transform(X).shape
```

```python
#get the name of those 7 columns
ct.get_feature_names_out()
```
