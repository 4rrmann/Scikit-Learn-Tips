*need **something better than SimpleImputer** for missing value imputation?*

>*try **KNNimputer** or **iterativeImputer**(inspired by R's MICE package),*

<br>*both are **multivarite approaches** (they take other features into account!)*

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')

X = df[['fare','age']].tail()
X
```

> ***iterative imputer:*** *Fills missing values by repeatedly predicting each feature from the remaining features using a learned model, refining estimates until stable values are reached.*

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
impute_it = IterativeImputer()
impute_it.fit_transform(X)
```

> ***KNNimputer:*** *Replaces missing values using the average (or weighted average) of the k-nearest samples based on feature similarity.*

```python
from sklearn.impute import KNNImputer
```

```python
impute_knn = KNNImputer(n_neighbors=2)
impute_knn.fit_transform(X)
```
