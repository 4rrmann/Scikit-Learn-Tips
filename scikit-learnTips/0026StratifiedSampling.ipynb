***`stratify=y`: class proportions are preserved when splitting***
<br>*especially important if you have class imbalance!*

```python
import pandas as pd

# Simple imbalanced dataset
data = {
    "amount": [120, 50, 300, 20, 15, 500, 10, 60],
    "transactions": [1, 2, 1, 5, 6, 1, 7, 3],
    "target": [
        "not fraud", "not fraud", "not fraud", "not fraud", "not fraud",
        "fraud", "fraud", "not fraud"
    ]
}

df = pd.DataFrame(data)
df
```

```python
X = df.drop('target', axis=1)
```

```python
y = df['target']
```

***not Stratified:***
> *`y_train` contains **NONE** of the minority class, whereas `y_test` contains **ALL** of the minority class. (This is bad mann)*

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.5, random_state=3)
```

```python
y_train
```

```python
y_test
```

***Stratified:***
> *class proportions are the **SAME** in `y_train` and `y_test`. (This is good)*

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.5, random_state=3, stratify=y)
```

```python
y_train
```

```python
y_test
```
