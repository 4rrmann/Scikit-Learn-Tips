*if you use cross-validation and your samples are NOT in an arbitary order, shuffling may be required to get meaningful results.*<br>
> ***use KFold or StratifiedKFold in order to shuffle!***

```python
from sklearn.model_selection import KFold, StratifiedGroupKFold, cross_val_score
```

>***use KFold with `regression` problems:***

```python
kf = KFold(5, shuffle=True, random_state=1)

cross_val_score(reg,   #regressor
                X_reg, #features
                y_reg, #labels
                cv=kf,
                scoring='r2')
```

> ***use StratifiedGroupKFold with `classification` problems:***

```python
skf = StratifiedGroupKFold(5, shuffle=True, random_state=1)

cross_val_score(clf,   #classifier
                X_clf, #features
                y_clf, #labels
                cv=skf,
                scoring='accuracy')
```
