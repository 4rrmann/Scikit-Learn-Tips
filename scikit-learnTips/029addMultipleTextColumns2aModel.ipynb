***want to vectorize two text columns in a ColumnTransformer?***
> *you can't pass them in a list, but you can **pass the vectorizer twice!***<br>
*(They'll learn separate vocabularies)*

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

sdf = df[['age', 'pclass', 'who', 'class', 'fare', 'survived']]
sdf = sdf.dropna()

X = sdf.drop('survived', axis=1)
y = sdf['survived']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
from sklearn.compose import make_column_transformer
ct = make_column_transformer((vect, 'who'), (vect, 'class'))
ct.fit_transform(X)
```
