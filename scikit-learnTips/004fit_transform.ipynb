> ***use `fit_transform` on training data, but `transform` (only) on testing/new data.***
- *applies the same transformations to both sets of data, which creates consistent columns and **prevents data leakage!***

> *Models (and transformers) must learn only from training data,
<br>then apply that learning to test/new data -**without learning again.***

- *The model **learns patterns/statistics from training data***
- *Those learned values are **reused on test data***
- ***Test data is never used to learn anything*** *(if it does, then it creates data leakage)*
