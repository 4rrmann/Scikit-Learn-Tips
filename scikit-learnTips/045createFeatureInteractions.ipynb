*want to include **featuree interactions** in your model? use **PolynomialFeatures!***<br>
> *This is impractical if you have lots of features, and unnecessary if you're using a tree-based model.*

```python
import pandas as pd
import numpy as np

X = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 4, 4],
    "C": [0, 10, 100],
})

X
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(include_bias=False, interaction_only=True)
```

```python
#output columns: A,B,C, A*B, A*C, B*C
poly.fit_transform(X)
```
