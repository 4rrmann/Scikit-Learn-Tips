*two common ways to encode categorical features:*
- ***OneHotEncoder** for unordered (nominal) data*
- ***OrdinalEncoder** for ordered (ordinal) data*

```python
import pandas as pd

df = pd.DataFrame({
    "Shape": ["square", "square", "oval", "circle"],
    "Class": ["third", "first", "second", "third"],
    "Size": ["S", "S", "L", "XL"]
})

X = df
X
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
#left-to-right column is alphabetical (Circle, Oval, Squre):
ohe = OneHotEncoder(sparse_output=False)
ohe.fit_transform(X[['Shape']])
#+1
```

```python
#category ordering (within each feature) is defined by you:
oe = OrdinalEncoder(categories=[['first','second','third'], ['S','M','L','XL']])
oe.fit_transform(X[['Class', 'Size']])
```

*LE and OE **almost works** similarly, except with LE you can't define the ordering and you can only encode one thing t a time where as in OE you can encode multiple things t a time.*
> ***LabelEncoder** is for labels, not features!*
