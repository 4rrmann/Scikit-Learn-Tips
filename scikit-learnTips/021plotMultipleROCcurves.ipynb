```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

sdf = df[['age', 'pclass', 'fare', 'survived']]
sdf = sdf.dropna()

X = sdf.drop('survived', axis=1)
y = sdf['survived']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import RocCurveDisplay
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
lr.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

> ***Plotting multiple ROC curves on the same plot lets you visually compare which model separates classes better across all thresholds, making it easy to pick the strongest overall performer.***

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt

# First model (creates the figure)
disp = RocCurveDisplay.from_estimator(
    lr,
    X_test,
    y_test,
)

# Other models on same axis
RocCurveDisplay.from_estimator(
    dt,
    X_test,
    y_test,
    ax=disp.ax_
)

RocCurveDisplay.from_estimator(
    rf,
    X_test,
    y_test,
    ax=disp.ax_
)

plt.show()
```
