*use `drop='if_binary` with OneHotEncoder to drop the first category ONLY if it's a binary features(meaning it has exactly two categories).*

```python
import pandas as pd

X = pd.DataFrame({
    "Shape": ["circle", "oval", "square", "square"],
    "Color": ["pink", "yellow", "pink", "yellow"]
})

X
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
#drop=None (default) creates one feature column per category:
X_ohe = ohe.fit_transform(X)
df_ohe = pd.DataFrame(
    X_ohe.toarray(),
    columns=ohe.get_feature_names_out()
)

df_ohe
```

```python
#drop='first' drops the first category in each feature:
ohe = OneHotEncoder(drop='first')

X_ohe = ohe.fit_transform(X)
df_ohe = pd.DataFrame(
    X_ohe.toarray(),
    columns=ohe.get_feature_names_out()
)

df_ohe
```

```python
#drop='if binary' drops the first category of binary features:
ohe = OneHotEncoder(drop='if_binary')
X_ohe = ohe.fit_transform(X)
df_ohe = pd.DataFrame(
    X_ohe.toarray(),
    columns=ohe.get_feature_names_out()
)

df_ohe
```
