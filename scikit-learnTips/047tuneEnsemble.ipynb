*want to improve the accuracy of your VotingClassifier? Try tuning the 'voting' and 'weights' parameters to change how predictions are combined!*<br>
***if you're using VotingRegressor, just tune the `weights` parameter***

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

sdf = df[['age', 'pclass', 'fare', 'survived']]
sdf = sdf.dropna()

X = sdf.drop('survived', axis=1)
y = sdf['survived']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
lr = LogisticRegression()
rf = RandomForestClassifier()
nb = MultinomialNB()
```

```python
#create an ensemble of 3 classifiers
vc = VotingClassifier([('clf1', lr), ('clf2', rf), ('clf3', nb)])
cross_val_score(vc, X,y).mean()
```

```python
#define VotingClassifier parameters to search
params = {'voting': ['hard', 'soft'],
          'weights': [(1,1,1), (2,1,1), (1,2,1), (1,1,2)]}
```

```python
#fine the best set of parameters
grid = GridSearchCV(vc, params)
grid.fit(X,y)
grid.best_params_
```

```python
#accuracy has improved!
grid.best_score_
```

>*it's highly likely that the accuracy of the Ensemble could be **improved even further by tuning each model within the ensemble***
